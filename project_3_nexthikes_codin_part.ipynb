# Detect and remove outliers

```python
data.describe()
```

```python
## show the distplot of "SalePrice" column.

sns.distplot(data['SalePrice'])
```

```python
sns.boxplot(data['SalePrice'])
```

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))
    return outliers

# Detect outliers in the 'SalePrice' column
outliers = detect_outliers_iqr(data[['SalePrice']])
print(outliers)
```

```python
# Remove outliers
data_no_outliers = data[~outliers.any(axis=1)]

# Display the shape of the dataframe before and after removing outliers
print("Original dataframe shape:", data.shape)
print("Dataframe shape after removing outliers:", data_no_outliers.shape)
```
